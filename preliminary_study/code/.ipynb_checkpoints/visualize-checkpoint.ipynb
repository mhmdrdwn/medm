```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/clean_data.csv', delimiter = ',')
df
```

```python
y= ['bubble_random', 'insertion_random']
x = 'list_length'
df.plot(x, y, figsize=(20,10), marker='o')
plt.xlabel('Data Size')
plt.ylabel('Run Time')
plt.show()
```

```python
y= ['bubble_sorted', 'insertion_sorted']
x = 'list_length'
df.plot(x, y, figsize=(20,10), marker='o')
plt.xlabel('Data Size')
plt.ylabel('Run Time')
plt.show()
```
